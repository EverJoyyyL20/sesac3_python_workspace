# 반복문을 이용해 파일 읽고 쓰기

## 파일에 문자열 한 줄씩 쓰기

```python
f=open('two_times_table.txt','w')
for num in range(1,6):
    format_string="2 x {0} ={1}\n".format(num,2*num)
    f.write(format_string)
f.close()
```

```python
f=open("two_times_table.txt")
line1=f.readline()
line2=f.readline()
f.close()
print(line1,end="")
print(line2,end="")
```

```python
f=open("two_times_table.txt")
line=f.readline()
while line:
    print(line,end="")
    line=f.readline()
f.close()
```

```python
f=open("two_times_table.txt")
lines=f.readlines()
f.close()

print(lines)
```

```python
f=open("two_times_table.txt")
for line in f.readline():
    print(line,end="")
f.close()
```

```python
f=open("two_times_table.txt")
for line in f:
    print(line,end="")
f.close()
```

```python
with open('myTextFile2.txt','w') as f:
    f.write('File read/write test2: line1\n')
    f.write('File read/write test2: line2\n')
    f.write('File read/write test2: line3\n')
```

```python
with open('myTextFile2.txt') as f:
    file_string=f.read()
    print(file_string)
```

```python
with open('myTextFile3.txt','w') as f:
    for num in range(1,6):
        format_string="3 x {0} ={1}\n".format(num,3*num)
        f.write(format_string)
```

```python
with open('myTextFile3.txt.','r') as f:
    for line in f:
        print(line,end="")
```

```python
a=5 #전역변수
def func1():
    a=1 #지역변수,func1()에서만 사용
    print("[func1] 지역변수 a=",a)
def func2():
    a=2 # 지역 변수,func2()에서만 사용
    print("[func2] 지역변수 a=",a)
def func3():
    print("[func3] 전역변수 a=",a)

def func4():
    global a
    a=4
    print("[func4] 전역변수 a=",a)
```

```python
func1() #함수 func1() 호출
func2() #함수 func2() 호출
print("전역 변수 a=",a) #전역변수 출력
```

```python
func3() #함수 func3()호출
func4() #함수 func4()호출
func3() #함수 func3() 호출
```
